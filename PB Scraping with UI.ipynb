```python
def popup_close(driver):
    close_attempt=2
    popup_close_status=False
    while(popup_close_status==False and close_attempt>0):
        try:
            print(1)
            driver.find_element(By.XPATH,"//a[contains(@class,'close ng-star-inserted')]").click()
            popup_close_status=True
            print('popup 1 close')                
        except:
            print(2)
            try:
                driver.find_element(By.XPATH,"//button[contains(@class,'callleter')]").click()
                popup_close_status=True
                print('popup 2 close')
            except:
                print(3)
                try:
                    driver.find_element(By.XPATH,"//input[contains(@class,'no btn lang')]").click()
                    popup_close_status=True
                    print('popup 3 close')
                except:
                    print(4)
                    try:
                        driver.find_element(By.XPATH,"//button[contains(@class,'call-to-action2')]").click()
                        popup_close_status=True
                        print('popup 4 close')
                    except:
                        print('No Valid Pop Found')
                        close_attempt=close_attempt-1
                        popup_close_status=False
                        print('Close Attempt - ',close_attempt)
            
```

```python
def edit_variables(driver,gender,dob,som,income,edu,emp,pin):  
    source=driver.page_source
    page_soup=soup(source,'html.parser')
    loading_boolen=False
    loading_attempt=5
    while(loading_boolen==False and loading_attempt>0):
        source=driver.page_source
        page_soup=soup(source,'html.parser')
        edit_page_loading_status=page_soup.find_all('div',{'class':'profile'})[0].find_all('li')[4].text
        print('edit_page_loading_status - ',edit_page_loading_status)
        try:
            if edit_page_loading_status=='Edit':
                driver.find_element(By.XPATH,"//a[text()='Edit']").click()
                gender_xpath="//span[text()='{gender}']".format(gender=gender)
                driver.find_element(By.XPATH,gender_xpath).click()
                driver.find_element(By.XPATH,"//input[contains(@placeholder,'DD/MM/YYYY')]").clear()
                driver.find_element(By.XPATH,"//input[contains(@placeholder,'DD/MM/YYYY')]").send_keys(dob)
                som_non_som_xpath="//span[text()='{som}']".format(som=som)
                driver.find_element(By.XPATH,som_non_som_xpath).click()
                temp_income=driver.find_element(By.NAME,"annualIncome")
                annual_income=Select(temp_income)
                annual_income.select_by_visible_text(income)
                temp_qua=driver.find_element(By.NAME,"educationalQualificationType")
                edu_qua=Select(temp_qua)
                edu_qua.select_by_visible_text(edu)
                emp_xpath="//span[text()='{emp}']".format(emp=emp)
                element=driver.find_element(By.XPATH,emp_xpath)
                driver.execute_script("arguments[0].click();", element)
                driver.find_element(By.XPATH,"//input[contains(@placeholder,'Enter Pin Code')]").send_keys(pin)
                driver.find_element(By.XPATH,"//span[text()='Save']").click()
                loading_boolen=True
                print('Edit Complete')
            else:
                loading_attempt=loading_attempt-1
                loading_boolen=False
                print('Page Not Loadded') 
        except:
            loading_boolen=False
            loading_attempt=loading_attempt-1
            print('Edit Attempt - ',loading_attempt)
```
